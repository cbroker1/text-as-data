# Setup

```python
import numpy as np
import pandas as pd

from collections import Counter
from sklearn.cluster import DBSCAN
import scipy.cluster.hierarchy as sch
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 14, 6
```

# Load Data

```python
df_game_reviews = pd.read_csv(r'../data/processed/game_reviews_processed.csv', low_memory=False)
```

```python
# clean
# subset data that's hand-coded
def prune_cols(df):
    # keep only needed cols and subset english only
    global df_game_reviews
    df_game_reviews = df_game_reviews.loc[df_game_reviews['Sentiment'].notnull(), ['Joy Intensity', 'Anger Intensity', 'Fear Intensity', 'Sadness Intensity', 'Sentiment']] # subsets for english
    return df_game_reviews

prune_cols(df_game_reviews)
```

# Prep Data For Model

```python
data = df_game_reviews.iloc[:,0:4]
target = df_game_reviews.iloc[:,4]
```

# PART 1: Build DBSCAN Model 

```python
# build model
model = DBSCAN (eps=1, min_samples=5).fit(data)
```

# Visualize Results

```python
outliers_df = pd.DataFrame(data)

print(Counter(model.labels_))

print(outliers_df[model.labels_==-1]) #-1 is outlier
```

```python
fig = plt.figure()

ax = fig.add_axes([.1,.1, 1, 1])

colors = model.labels_

ax.scatter(data.iloc[:,0].values, data.iloc[:,1].values, c=colors, s=120)

ax.set_xlabel('Joy')
ax.set_ylabel('Anger')

plt.title('Joy vs Anger')
```

# PART 2: Build Hierarchical Model

```python
X = np.array(X)
```

```python
# create a dendrogram
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

# 1. Fit Hierarchical model to your project data. Pick three cut heights based on visual interpretation of the dendrogram. Do the clusters make sense?

```python
# 4 clusters
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')

y_hc = hc.fit_predict(X)

# visualize clustering
plt.scatter(X[y_hc ==0,0], X[y_hc == 0,1], s=120, c='cyan')
plt.scatter(X[y_hc ==1,0], X[y_hc == 1,1], s=120, c='yellow')
plt.scatter(X[y_hc ==2,0], X[y_hc == 2,1], s=120, c='red')
plt.scatter(X[y_hc ==3,0], X[y_hc == 3,1], s=120, c='green')
```

```python
# 3 clusters
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')

y_hc = hc.fit_predict(X)

# visualize clustering
plt.scatter(X[y_hc ==0,0], X[y_hc == 0,1], s=120, c='cyan')
plt.scatter(X[y_hc ==1,0], X[y_hc == 1,1], s=120, c='yellow')
plt.scatter(X[y_hc ==2,0], X[y_hc == 2,1], s=120, c='red')
```

```python
# 2 clusters
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')

y_hc = hc.fit_predict(X)

# visualize clustering
plt.scatter(X[y_hc ==0,0], X[y_hc == 0,1], s=120, c='cyan')
plt.scatter(X[y_hc ==1,0], X[y_hc == 1,1], s=120, c='yellow')
```

I think the clusters of '3' makes the most sense, given the sentiment labels pos, neg, and neutral.

# 2. Fit Hierarchical model using the three different distance measures. Compare the dendograms and results for the same cut heights.

According to scipy documentation, here are the available distance measures: https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html#scipy.cluster.hierarchy.linkage

#### Single

```python
# single 
dendrogram = sch.dendrogram(sch.linkage(X, method='single'))
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
y_hc = hc.fit_predict(X)
```

```python
# visualize clustering
plt.scatter(X[y_hc ==0,0], X[y_hc == 0,1], s=120, c='cyan')
plt.scatter(X[y_hc ==1,0], X[y_hc == 1,1], s=120, c='yellow')
plt.scatter(X[y_hc ==2,0], X[y_hc == 2,1], s=120, c='red')
```

#### Complete

```python
# complete
dendrogram = sch.dendrogram(sch.linkage(X, method='complete'))
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
y_hc = hc.fit_predict(X)
```

```python
# visualize clustering
plt.scatter(X[y_hc ==0,0], X[y_hc == 0,1], s=120, c='cyan')
plt.scatter(X[y_hc ==1,0], X[y_hc == 1,1], s=120, c='yellow')
plt.scatter(X[y_hc ==2,0], X[y_hc == 2,1], s=120, c='red')
```

#### Average

```python
# complete
dendrogram = sch.dendrogram(sch.linkage(X, method='average'))
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')
y_hc = hc.fit_predict(X)
```

```python
# visualize clustering
plt.scatter(X[y_hc ==0,0], X[y_hc == 0,1], s=120, c='cyan')
plt.scatter(X[y_hc ==1,0], X[y_hc == 1,1], s=120, c='yellow')
plt.scatter(X[y_hc ==2,0], X[y_hc == 2,1], s=120, c='red')
```
