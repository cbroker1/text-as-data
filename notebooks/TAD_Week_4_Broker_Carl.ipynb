# 1. Which Tweets from Nuclear Sentiment dataset mention Japan.

```python
# mention japan function
def mention_japan(x):
    x = pd.read_csv(x)
    pattern = 'Japan'# set pattern
    output = x[x.tweet_text.str.contains(pattern, case=False)] #subset df w/tweets containing pattern
    return output

mention_japan('sentiment_nuclear_power _1_.csv')
```

# 2. Find all words starting with an upper case in Supreme Court case from Week 3 using RegEx.

```python
# function that counts upper case items in a pdf
def upp_func(x):
    upp_count = 0
    pdfFileObj = open(x, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    l = pdfReader.numPages
    
    for item in range(0,l,1):
        pageObj = pdfReader.getPage(item)
        temp_doc = pageObj.extractText()
        temp_doc = temp_doc.replace('\n', '')
        #print(temp_doc)
        for word in temp_doc:
            if word[0].isupper() == True:
                upp_count += 1
    return upp_count    
```

# 3. Remove all occurrences of Twitter usernames (@<username>) and hashtags (#<hashtag>) from the Nuclear Sentiment dataset from Week 3

```python
# clean twitter csv of '@' and '#' containing substrings

def regex_magic(tweet):
    tweet = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split())
    return tweet

def clean_twitter(csv):
    df_nuke = pd.read_csv(csv)
    df_nuke['tweet_text'] = df_nuke['tweet_text'].apply(regex_magic)
    return df_nuke
    
clean_twitter('sentiment_nuclear_power _1_.csv')
```
